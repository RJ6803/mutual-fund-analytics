# DAY 4 — Fund Performance Analytics

### IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import linregress
```

### LOAD THE DATASET

```python
nav = pd.read_csv("../data/processed/nav_history_clean.csv")

performance = pd.read_csv("../data/processed/scheme_performance_clean.csv")

benchmark = pd.read_csv("../data/processed/benchmark_indices_clean.csv")

fund_master = pd.read_csv("../data/processed/fund_master_clean.csv")
```

### 1.Daily Returns

```python
nav["date"] = pd.to_datetime(nav["date"])

nav = nav.sort_values(
    ["amfi_code", "date"]
)

nav["daily_return"] = (
    nav.groupby("amfi_code")["nav"]
       .pct_change()
)

nav.head()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    nav["daily_return"].dropna(),
    bins=50,
    kde=True
)

plt.title("Daily Return Distribution")

plt.show()
```

### 2.CAGR

```python
cagr_results = []

for code in nav["amfi_code"].unique():

    df = nav[nav["amfi_code"] == code]

    start_nav = df.iloc[0]["nav"]
    end_nav = df.iloc[-1]["nav"]

    years = (
        (df.iloc[-1]["date"] - df.iloc[0]["date"]).days
        / 365
    )

    cagr = (end_nav / start_nav) ** (1 / years) - 1

    cagr_results.append([code, cagr])

cagr_df = pd.DataFrame(
    cagr_results,
    columns=["amfi_code", "cagr"]
)

cagr_df.head()
```

### 3.Sharpe Ratio

```python
RF = 0.065
```

```python
sharpe_list = []

for code in nav["amfi_code"].unique():

    r = nav[
        nav["amfi_code"] == code
    ]["daily_return"].dropna()

    mean_return = r.mean() * 252

    std_return = r.std()

    sharpe = (
        (mean_return - RF)
        / std_return
    ) * np.sqrt(252)

    sharpe_list.append(
        [code, sharpe]
    )

sharpe_df = pd.DataFrame(
    sharpe_list,
    columns=["amfi_code", "sharpe_ratio"]
)

sharpe_df.sort_values(
    "sharpe_ratio",
    ascending=False
).head()
```

### 4.Sortino Ratio

```python
sortino_list = []

for code in nav["amfi_code"].unique():

    r = nav[
        nav["amfi_code"] == code
    ]["daily_return"].dropna()

    downside = r[r < 0]

    downside_std = downside.std()

    mean_return = r.mean() * 252

    sortino = (
        (mean_return - RF)
        / downside_std
    ) * np.sqrt(252)

    sortino_list.append(
        [code, sortino]
    )

sortino_df = pd.DataFrame(
    sortino_list,
    columns=[
        "amfi_code",
        "sortino_ratio"
    ]
)
sortino_df
```

```python
print(nav["date"].dtype)
print(benchmark["date"].dtype)
```

```python
alpha_beta = []

for code in nav["amfi_code"].unique():

    fund = nav[
        nav["amfi_code"] == code
    ][["date", "daily_return"]]

    merged = pd.merge(
        fund,
        benchmark[
            ["date", "return"]
        ],
        on="date"
    )

    merged = merged.dropna()

    slope, intercept, r, p, se = linregress(
        merged["return"],
        merged["daily_return"]
    )

    beta = slope

    alpha = intercept * 252

    alpha_beta.append(
        [code, alpha, beta]
    )

alpha_beta_df = pd.DataFrame(
    alpha_beta,
    columns=[
        "amfi_code",
        "alpha",
        "beta"
    ]
)

alpha_beta_df
```

```python
alpha_beta_df.to_csv(
    "../data/processed/alpha_beta.csv",
    index=False
)
```

### 6.Maximum Drawdown

```python
dd_list = []

for code in nav["amfi_code"].unique():

    df = nav[
        nav["amfi_code"] == code
    ].copy()

    df["running_max"] = (
        df["nav"].cummax()
    )

    df["drawdown"] = (
        df["nav"]
        / df["running_max"]
        - 1
    )

    max_dd = df["drawdown"].min()

    dd_list.append(
        [code, max_dd]
    )

dd_df = pd.DataFrame(
    dd_list,
    columns=[
        "amfi_code",
        "max_drawdown"
    ]
)
dd_df
```

### 7.Fund Scorecard

```python
scorecard = (
    cagr_df
    .merge(sharpe_df)
    .merge(alpha_beta_df)
    .merge(dd_df)
)
scorecard
```

```python
scorecard["return_rank"] = (
    scorecard["cagr"]
    .rank(pct=True)
)
scorecard["return_rank"]

scorecard["sharpe_rank"] = (
    scorecard["sharpe_ratio"]
    .rank(pct=True)
)
scorecard["sharpe_rank"]

scorecard["alpha_rank"] = (
    scorecard["alpha"]
    .rank(pct=True)
)
scorecard["alpha_rank"]

scorecard["dd_rank"] = (
    (-scorecard["max_drawdown"])
    .rank(pct=True)
)
scorecard["dd_rank"]
```

```python
scorecard["fund_score"] = (
    scorecard["return_rank"] * 30 +
    scorecard["sharpe_rank"] * 25 +
    scorecard["alpha_rank"] * 20 +
    scorecard["dd_rank"] * 10
)
scorecard["fund_score"] 
```

```python
scorecard["fund_score"] = (
    scorecard["fund_score"]
    / scorecard["fund_score"].max()
) * 100
scorecard["fund_score"]
```

```python
scorecard.to_csv(
    "../data/processed/fund_scorecard.csv",
    index=False
)
```
