#### IMPORT PYTHON LIBRARY

```python
import pandas as pd
import numpy as np
import requests
import os
```

#### FETCH THE DATASET

```python
import os

data_folder = "../data/raw"

csv_files = sorted(
    [f for f in os.listdir(data_folder)
     if f.endswith(".csv")
     ]
)

csv_files
```

#### LOAD THE DATASET

```python
datasets = {}

for file in csv_files:
    path = os.path.join(data_folder, file)
    datasets[file] = pd.read_csv(path)

print("All datasets loaded successfully")
```

#### print each dataset shape, dtype, head

```python
for name, df in datasets.items():
    print("\n\n")
    print("=" * 60)
    print(name)

    print("\nShape:")
    print(df.shape)

    print("\nData Types:")
    print(df.dtypes)

    print("\nHead:")
    print(df.head())

    print("\nMissing Values:")
    print(df.isnull().sum())
```

#### any anomalies
- Missing values
- Duplicate rows
- Wrong data types
- Empty columns

```python
for name, df in datasets.items():

    print("\n", name)

    print("Duplicates:", df.duplicated().sum())

    print("Missing Values:",
          df.isnull().sum().sum())
```

#### Fetch live NAV

```python
url = "https://api.mfapi.in/mf/125497"

response = requests.get(url)

data = response.json()

print(data["meta"]["scheme_name"])
```

#### convert NAV into dataframe and save it

```python
nav_df = pd.DataFrame(data["data"])

nav_df.to_csv(
    "../data/raw/HDFC_Top100_Direct.csv",
    index=False
)

print(nav_df.shape)
print(nav_df.dtypes)
print(nav_df.head())
```

#### Fetch NAV for 5 key schemes

```python
schemes = {
    "SBI_Bluechip":119551,
    "ICICI_Bluechip":120503,
    "Nippon_LargeCap":118632,
    "Axis_Bluechip":119092,
    "Kotak_Bluechip":120841
}
```

#### download all NAV

```python
for name, code in schemes.items():

    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)

    data = response.json()

    df = pd.DataFrame(data["data"])

    df.to_csv(
        f"../data/raw/{name}.csv",
        index=False
    )

    print(f"{name} saved")
```
